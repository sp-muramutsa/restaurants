```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from time import sleep
import csv
```

```python
driver = webdriver.Chrome()
mcdonalds_url = "https://www.mcdonalds.com/us/en-us/restaurant-locator.html?"
driver.get(mcdonalds_url)
sleep(10)
```

```python
input_formula = driver.find_element("name", "search-val")
location = input("Type Zip or City State: ")
input_formula.send_keys(location)
input_formula.send_keys(Keys.ENTER)
```

```python
# Load all restaurants
while True:
    try:
        load_more_button = driver.find_element(By.CLASS_NAME, "cmp-restaurant-locator__restaurant-load-more-btn")
        button_element = load_more_button.find_element(By.CLASS_NAME, "cmp-button")
        if button_element.is_displayed():
            ActionChains(driver).move_to_element(button_element).click().perform()
        else:
            break  
    except Exception as e:
        print(f"An error occurred: {e}")
        break

    
# Collect data from each restaurant   
restaurant_divs = driver.find_elements(By.CSS_SELECTOR, ".cmp-restaurant-locator__restaurant-list-item")
restaurants = []
for div in restaurant_divs:
    name = div.find_element(By.TAG_NAME, "a").text
    address = div.find_element(By.CSS_SELECTOR, ".cmp-restaurant-locator__restaurant-list-item-details-al2").text
    
    # Collect data for just the McDonald's file
    restaurant = {"name": name, "address": address}
    restaurants.append(restaurant)
```
