```python
import pandas as pd
import folium

df = pd.read_csv("restaurants_geo.csv", encoding="utf-8")
m = folium.Map(location=[32.7767, -96.7970], tiles="OpenStreetMap", zoom_start=11)

# Apply conversion to string and filter rows
for i, row in df.iterrows():
    latitude = df.at[i, "lat"]
    longitude = df.at[i, "long"]
    
    restaurant = df.at[i,"restaurant"]
    popup_text = (str(df.at[i, "restaurant"]) + "<br>" + str(df.at[i,"street"])
                    + "<br>" + str(df.at[i,"zip"]) + ' ' + str(df.at[i, "city"]))
    
    match restaurant:
        case "Subway":
            color= "orange"
        case "Starbucks":
            color = "green"
        case _ :
            color = "red"

    folium.Marker(location=[latitude, longitude],popup=popup_text, icon=folium.Icon(color=color)).add_to(m)
m
```

```python
m.save("restaurants.html")
```
